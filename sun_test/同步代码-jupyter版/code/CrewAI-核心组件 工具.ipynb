```python
import os
from crewai import Agent, Task, Crew
from dotenv import load_dotenv
import os
# åŠ è½½ .env æ–‡ä»¶
load_dotenv()


from crewai_tools import (
    DirectoryReadTool,
    FileReadTool,
    SerperDevTool,
    WebsiteSearchTool
)

# è®¾ç½® API å¯†é’¥
os.environ["SERPER_API_KEY"] = os.environ.get("SERPER_API_KEY")
os.environ["OPENAI_API_KEY"] = os.environ.get("OPENAI_API_KEY")
os.environ["OPENAI_API_BASE"] = os.environ.get("OPENAI_API_BASE")  # 

# å®ä¾‹åŒ–å·¥å…·
docs_tool = DirectoryReadTool(directory='./blog-posts')
file_tool = FileReadTool()
search_tool = SerperDevTool()
web_rag_tool = WebsiteSearchTool()

# åˆ›å»ºæ™ºèƒ½ä½“
researcher = Agent(
    role='å¸‚åœºç ”ç©¶åˆ†æå¸ˆ',
    goal='æä¾›äººå·¥æ™ºèƒ½è¡Œä¸šçš„æœ€æ–°å¸‚åœºåˆ†æ',
    backstory='ä¸€ä½å¯¹å¸‚åœºè¶‹åŠ¿æœ‰æ•é”æ´å¯ŸåŠ›çš„ä¸“ä¸šåˆ†æå¸ˆã€‚',
    tools=[search_tool, web_rag_tool],
    verbose=True
)

writer = Agent(
    role='å†…å®¹æ’°å†™å‘˜',
    goal='æ’°å†™å…³äºäººå·¥æ™ºèƒ½è¡Œä¸šçš„å¼•äººå…¥èƒœçš„åšå®¢æ–‡ç« ',
    backstory='ä¸€ä½çƒ­çˆ±æŠ€æœ¯çš„ç†Ÿç»ƒä½œå®¶ã€‚',
    tools=[docs_tool, file_tool],
    verbose=True
)

# å®šä¹‰ä»»åŠ¡
research = Task(
    description='ç ”ç©¶äººå·¥æ™ºèƒ½è¡Œä¸šçš„æœ€æ–°è¶‹åŠ¿å¹¶æä¾›æ‘˜è¦ã€‚',
    expected_output='äººå·¥æ™ºèƒ½è¡Œä¸šå‰3ä¸ªçƒ­é—¨å‘å±•è¶‹åŠ¿çš„æ‘˜è¦ï¼Œå¹¶å¯¹å…¶é‡è¦æ€§æä¾›ç‹¬ç‰¹è§è§£ã€‚',
    agent=researcher
)

write = Task(
    description='æ ¹æ®ç ”ç©¶åˆ†æå¸ˆçš„æ‘˜è¦ï¼Œæ’°å†™ä¸€ç¯‡å…³äºäººå·¥æ™ºèƒ½è¡Œä¸šçš„å¼•äººå…¥èƒœçš„åšå®¢æ–‡ç« ã€‚ä»ç›®å½•ä¸­çš„æœ€æ–°åšå®¢æ–‡ç« ä¸­æ±²å–çµæ„Ÿã€‚',
    expected_output='ä¸€ç¯‡ä»¥markdownæ ¼å¼ç¼–å†™çš„4æ®µè½åšå®¢æ–‡ç« ï¼Œå†…å®¹å¼•äººå…¥èƒœã€ä¿¡æ¯ä¸°å¯Œä¸”æ˜“äºç†è§£ï¼Œé¿å…å¤æ‚æœ¯è¯­ã€‚',
    agent=writer,
    output_file='blog-posts/new_post.md'  # æœ€ç»ˆçš„åšå®¢æ–‡ç« å°†ä¿å­˜åœ¨è¿™é‡Œ
)

# ç»„å»ºä¸€ä¸ªå¯ç”¨è§„åˆ’åŠŸèƒ½çš„å›¢é˜Ÿ
crew = Crew(
    agents=[researcher, writer],
    tasks=[research, write],
    verbose=True,
    planning=True,  # å¯ç”¨è§„åˆ’åŠŸèƒ½
)

# æ‰§è¡Œä»»åŠ¡
crew.kickoff()
```

```python
import os
from dotenv import load_dotenv
from crewai import Agent, Task, Crew
from crewai.tools import BaseTool
from pydantic import Field
from langchain_community.utilities.tavily_search import TavilySearchAPIWrapper

load_dotenv()

# Create the wrapper with just the API key
search = TavilySearchAPIWrapper(tavily_api_key=os.environ.get("TAVILY_API_KEY"))


from typing import ClassVar

class SearchTool(BaseTool):
    name: str = "Search"
    description: str = "Useful for search-based queries. Use this to find current information about markets, companies, and trends."
    search: ClassVar[TavilySearchAPIWrapper] = search

    def _run(self, query: str) -> str:
        """Execute the search query and return results"""
        try:
            return self.search.run(query)
        except Exception as e:
            return f"Error performing search: {str(e)}"

# Create Agents
researcher = Agent(
    role='Research Analyst',
    goal='Gather current market data and trends',
    backstory="""You are an expert research analyst with years of experience in
    gathering market intelligence. You're known for your ability to find
    relevant and up-to-date market information and present it in a clear,
    actionable format.""",
    tools=[SearchTool()],
    verbose=True
)
```
